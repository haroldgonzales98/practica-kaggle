```python
# AJUSTES Y DECLARACIONES VARIAS

personality_traits = ["EXT", "AGR", "CSN", "EST", "OPN"]
answer_columns = [trait + str(number) for trait in personality_traits for number in range(1, 11)]
answer_columns_time = [trait + str(number) + "_E" for trait in personality_traits for number in range(1, 11)]

# De acuerdo con el libro de códigos, las latitudes y longitudes dadas (columnas lat_appx_lots_of_err
# y long_appx_lots_of_err) son muy inexactas, por lo que se eliminarán ambas columnas.
data.drop(["lat_appx_lots_of_err", "long_appx_lots_of_err"], axis = 1, inplace = True) 

# Para una mayor precisión y simpleza, pasamos los tipos de respuesta a integers
data[answer_columns] = data[answer_columns].astype(int)

# Tabla auxiliar para mirar el count de las respuestas
# answer_table = data[answer_columns].apply(pd.Series.value_counts)
# print('Tamaño de la nueva tabla: ', answer_table.shape)
data[answer_columns].apply(pd.Series.value_counts)
data = data[(data[answer_columns] != 0).all(axis = 1)]

# Añadimos los tiempos gastados en cada respuesta
data[answer_columns_time] = data[answer_columns_time].apply(lambda x: x / 1000)
print('Valores nuevos de tiempos: ')
data[answer_columns_time].describe()
```

```python
# Algunos tiempos son excesivos o hay algunos negativos, por lo que si aceptamos un tiempo maximo
# de 30 segundos y un mínimo de 0, nos daría el siguiente resultado:

data_real = data[((data[answer_columns_time] < 30) & (data[answer_columns_time] > 0)).all(axis = 1)]
print('Tamaño de un dataset más ajustado al tiempo requerido: ', data_real.shape)
```

```python
# EDA: Calculo de tiempo por respuesta (data_real = tiempo menor a 30 segundos)

data_response = pd.melt(data_real[answer_columns_time])
data_response["trait"] = data_response["variable"].str.slice(0, 3)
```

```python
fig, axs = plt.subplots(ncols = 2, nrows = 3, figsize = (18, 18))

sns.boxplot(x = "variable", y = "value", data = data_response[data_response["trait"] == "EXT"], 
            showfliers = False, ax = axs[0, 0]).set_title("Extroversion")
sns.boxplot(x = "variable", y = "value", data = data_response[data_response["trait"] == "AGR"], 
            showfliers = False, ax = axs[0, 1]).set_title("Agreeableness")
sns.boxplot(x = "variable", y = "value", data = data_response[data_response["trait"] == "CSN"], 
            showfliers = False, ax = axs[1, 0]).set_title("Conscientiousness")
sns.boxplot(x = "variable", y = "value", data = data_response[data_response["trait"] == "EST"], 
            showfliers = False, ax = axs[1, 1]).set_title("Neuroticism")
sns.boxplot(x = "variable", y = "value", data = data_response[data_response["trait"] == "OPN"], 
            showfliers = False, ax = axs[2, 0]).set_title("Openness")

fig.delaxes(axs[2, 1])

for ax in axs.flat:
    ax.set(xlabel = None, ylabel = "Response time (seconds)")
```

Podemos comprobar que el tiempo de respuesta de cada pregunta es prácticamente la misma, salvo una (EXT1: I am the life of the party). Podemos deducir que se tarda en responder porque es la primera pregunta. Algunas respuestas son ligeramente más altas, probablemente, porque hubo problemas de comprensión lectora o porque la pregunta era más larga que el promedio de las demás.

Ahora vamos a hallar la correlación que pueda existir entre los rasgos de personalidad:

```python
#Hacemos a mano para obtener una puntuación para cada tratamiento de personalidad (los simbolos negativos y positivos
#indican que son respuestas a favor o, en cambio, respuestas en contra)

data["EXT"] = data["EXT1"] - data["EXT2"] + data["EXT3"] - data["EXT4"] + \
    data["EXT5"] - data["EXT6"] + data["EXT7"] - data["EXT8"] + data["EXT9"] - data["EXT10"]

data["EST"] = data["EST1"] - data["EST2"] + data["EST3"] - data["EST4"] + \
    data["EST5"] + data["EST6"] + data["EST7"] + data["EST8"] + data["EST9"] + data["EST10"]

data["AGR"] = - data["AGR1"] + data["AGR2"] - data["AGR3"] + data["AGR4"] - \
    data["AGR5"] + data["AGR6"] - data["AGR7"] + data["AGR8"] + data["AGR9"] + data["AGR10"]

data["CSN"] = data["CSN1"] - data["CSN2"] + data["CSN3"] - data["CSN4"] + \
    data["CSN5"] - data["CSN6"] + data["CSN7"] - data["CSN8"] + data["CSN9"] + data["CSN10"]

data["OPN"] = data["OPN1"] - data["OPN2"] + data["OPN3"] - data["OPN4"] + \
    data["OPN5"] - data["OPN6"] + data["OPN7"] - data["OPN8"] + data["OPN9"] + data["OPN10"]

correlation = data[personality_traits].corr()
plt.figure(figsize=(10, 10))
ax = sns.heatmap(correlation, annot=True)
```

Podemos comprobar que EXT (Extroversion) y AGR (Agreeableness) tienen un valor más alto que los demás, por tanto podemos decir que tienen mayor correlación. Sin embargo, podemos visualizar valores negativos de EST (Neuroticism) con EXT (-0.22) y CSN (Conscientiousness, -0.23).

## Model Learning

```python
# Primero entrenaremos con KMeans
# Intentaremos hallar cuántos clusters necesitamos

X=data.copy()
X.drop(X.columns[50:107], axis=1, inplace=True)
X.drop(X.columns[51:], axis=1, inplace=True)

df_model = X.drop('country', axis=1)

kmeans = KMeans(n_clusters=6)
k_fit = kmeans.fit(df_model)

pd.options.display.max_columns = 10
predictions = k_fit.labels_
df_model['Clusters'] = predictions
df_model.head()
```

```python
col_list = list(df_model)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]

data_sums = pd.DataFrame()
data_sums['extroversion'] = df_model[ext].sum(axis=1)/10
data_sums['neurotic'] = df_model[est].sum(axis=1)/10
data_sums['agreeable'] = df_model[agr].sum(axis=1)/10
data_sums['conscientious'] = df_model[csn].sum(axis=1)/10
data_sums['open'] = df_model[opn].sum(axis=1)/10
data_sums['clusters'] = predictions
data_sums.groupby('clusters').mean()
```

```python
pca = PCA(n_components=2)
pca_fit = pca.fit_transform(df_model)

df_pca = pd.DataFrame(data=pca_fit, columns=['PCA1', 'PCA2'])
df_pca['Clusters'] = predictions
plt.figure(figsize=(10,10))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='Clusters', palette='tab10', alpha=0.8)
plt.title('Personality Clusters after PCA');
```
