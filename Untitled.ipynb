```python
# IMPORT LIBRERIAS

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pycountry
```

```python
# lOADING DATASET Y ANALISIS

dataset = pd.read_csv('Base de Datos/data-final.csv', sep='\t')

print('Número de participantes: ', len(dataset))
print('Número de atributos: ', len(dataset.columns))
dataset.head()
```

```python
# ELIMINACION DE ATRIBUTOS INNECESARIOS

data = dataset.copy()
pd.options.display.max_columns = 150

data.drop(data.columns[50:107], axis=1, inplace=True)
data.drop(data.columns[51:], axis=1, inplace=True)

print('Número de atributos: ', len(data.columns))
data.head()
```

```python
# BUSQUEDA DE VALORES NULOS

print('¿Hay valores nulos? ', data.isnull().values.any())
print('Si es así, ¿cuántos valores nulos existen? ', data.isnull().values.sum())
print('')
print('* PROCESO DE ELIMINACION DE VALORES NULOS *')
data = dataset.dropna()
print('Nuevo valor de participantes después de la eliminación: ', len(data))
```

```python
# ANALISIS DE NUESTRA BASE DE DATOS

countries = pd.DataFrame(data['country'].value_counts())
countries_5000 = countries[countries['country'] >= 5000]
plt.figure(figsize=(15,5))
sns.barplot(data=countries_5000, x=countries_5000.index, y='country')
plt.title('Paises con más de 5000 participantes')
plt.ylabel('Participantes');
```

```python
# AJUSTES Y DECLARACIONES VARIAS

personality_traits = ["EXT", "AGR", "CSN", "EST", "OPN"]
answer_columns = [trait + str(number) for trait in personality_traits for number in range(1, 11)]
answer_columns_time = [trait + str(number) + "_E" for trait in personality_traits for number in range(1, 11)]

# De acuerdo con el libro de códigos, las latitudes y longitudes dadas (columnas lat_appx_lots_of_err
# y long_appx_lots_of_err) son muy inexactas, por lo que se eliminarán ambas columnas.
data.drop(["lat_appx_lots_of_err", "long_appx_lots_of_err"], axis = 1, inplace = True) 

# Para una mayor precisión y simpleza, pasamos los tipos de respuesta a integers
data[answer_columns] = data[answer_columns].astype(int)

# Tabla auxiliar para mirar el count de las respuestas
# answer_table = data[answer_columns].apply(pd.Series.value_counts)
# print('Tamaño de la nueva tabla: ', answer_table.shape)
data[answer_columns].apply(pd.Series.value_counts)
data = data[(data[answer_columns] != 0).all(axis = 1)]

# Añadimos los tiempos gastados en cada respuesta
data[answer_columns_time] = data[answer_columns_time].apply(lambda x: x / 1000)
print('Valores nuevos de tiempos: ')
data[answer_columns_time].describe()
```

```python
# Algunos tiempos son excesivos o hay algunos negativos, por lo que si aceptamos un tiempo maximo
# de 30 segundos y un mínimo de 0, nos daría el siguiente resultado:

data_real = data[((data[answer_columns_time] < 30) & (data[answer_columns_time] > 0)).all(axis = 1)]
print('Tamaño de un dataset más ajustado al tiempo requerido: ', data_real.shape)
```
